Import Library

```python
# memanggil library 
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert

import numpy as np
```

Percobaan 1 - Cropping Image

```python
# membuat gambar gambar
RandomImg = plt.imread("Random.jpg")
ShikamaruImg = plt.imread("Shikamaru.jpeg")

# menyalin lalu mengcropping gambar
RandomCropped = RandomImg.copy()
RandomCropped = RandomCropped[0:256, 120:420]

ShikamaruCropped = ShikamaruImg.copy()
ShikamaruCropped = ShikamaruCropped[64:256, 128:320]

print('Random Ori Shape : ',RandomImg.shape)
print('Random Crop Shape : ',RandomCropped.shape)

print('Shikamaru Ori Shape : ',ShikamaruImg.shape)
print('Shikamaru Crop Shape : ',ShikamaruCropped.shape)


fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(RandomImg)
ax[0].set_title("Citra Input 1")

ax[1].imshow(ShikamaruImg, cmap='gray')
ax[1].set_title('Citra Input 2')

ax[2].imshow(RandomCropped)
ax[2].set_title("Citra Output 1")

ax[3].imshow(ShikamaruCropped, cmap='gray')
ax[3].set_title('Citra Output 2')
```

Percobaan 2 - Citra Negative

```python
# melakukan invert pada citra awan yg sudah di cropped
inv = invert(RandomCropped)
print('Shape Input : ', RandomCropped.shape)
print('Shape Output : ',inv.shape)

# membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
ax = axes.ravel()
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(RandomCropped)
ax[0].set_title("Citra Input")

ax[1].hist(RandomCropped.ravel(), bins=420)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=420)
ax[3].set_title('Histogram Output')
```

```python
# mengubah citra menjadi grayscale
imggray = rgb2gray(ShikamaruCropped) 
print('img2gray shape = ',imggray.shape) 

# membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12
fig, axes = plt.subplots(1, 2, figsize=(12, 12)) 
ax = axes.ravel()

ax[0].imshow(ShikamaruCropped)
ax[0].set_title("img2 RGB Image")
ax[1].imshow(imggray,cmap=plt.cm.gray)
ax[1].set_title("img2 Grayscale Image")

# mengatur tata letak subplot
fig.tight_layout() 
# menampilkan subplot
plt.show() 
```

```python
# menyalin citra grayscale dan mengubah type datanya
copyShikamaru = imggray.copy().astype(float)

# untuk mendapatkan dimensi citra yang disalin
m1,n1 = copyShikamaru.shape

# membuat array kosong
output1 = np.empty([m1, n1])

for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        a1 = baris
        b1 = kolom
        output1[a1, b1] = copyShikamaru[baris, kolom] + 10

# membuat subplot dengan 2 baris dan 2 kolom dengan jarak 12        
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(copyShikamaru, cmap='gray')
ax[0].set_title("Citra Input")

ax[1].hist(copyShikamaru.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(output1, cmap='gray')
ax[2].set_title('Citra Output (Brightnes)')

ax[3].hist(output1.ravel(), bins=256)
ax[3].set_title('Histogram Input')
```
